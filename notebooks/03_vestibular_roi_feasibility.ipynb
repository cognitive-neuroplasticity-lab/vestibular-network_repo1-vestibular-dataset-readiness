```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
roi_df = pd.DataFrame({
    "ROI": [
        "Temporo-parietal junction (TPJ)",
        "Posterior insula",
        "Inferior parietal lobule",
        "Precuneus",
        "Hippocampus",
        "Parahippocampal cortex",
        "Cerebellum (vestibular zones)"
    ],
    "Primary Role": [
        "Multisensory integration",
        "Core vestibular cortical processing",
        "Body orientation / spatial transformation",
        "Self-location / imagery",
        "Navigation / spatial memory",
        "Scene context representation",
        "Balance / sensorimotor integration"
    ]
})

roi_df
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=roi_df, x="ROI", y=[1]*len(roi_df))
plt.title("Candidate Vestibular-Network Regions")
plt.xlabel("")
plt.ylabel("")
plt.xticks(rotation=35, ha="right")
plt.yticks([])
plt.tight_layout()
plt.show()
```
